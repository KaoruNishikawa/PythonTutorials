#### 標準ライブラリや外部ライブラリのインポート

既存の Python ライブラリを使う場合、そのライブラリを読み込む必要がある (インポート)。

インポートの書き方は2種類ある。

- `import matplotlib.pyplot`  
  → `matplotlib.pyplot.<使いたい機能>` でアクセスできる
- `from matplotlib import pyplot`  
  → `pyplot.<使いたい機能>` でアクセスできる

また、インポートしたライブラリに別の名前をつけて使うこともできる。

- `import matplotlib.pyplot as plt`  
- `from matplotlib import pyplot as plt`  
  → `plt.<使いたい機能>` でアクセスできる

````python
import matplotlib.pyplot as plt
````

#### 機能の探し方

機能を探すときには関連するワードで検索する他に、パッケージやオブジェクトに用意されている属性 (attribute) や関数 (method) の名前を `dir` 関数で一覧することもできる。

````python
obj = "python object"

print("obj の型 :", type(obj))
print("obj に用意されている属性・関数 :", dir(obj))
````

この一覧にはアンダーバーから始まるものが多くある。これらは Python の基本的な演算など内部的な処理のために定義された属性・関数で、ユーザーが使うことは想定されていない。そのため、これらを除いた一覧から使いたい機能を探せば十分である。

````python
print([attr for attr in dir(obj) if not attr.startswith("_")])
````

このようにして見つけた属性や関数は `<パッケージ名>.<モジュール名>.<関数・属性名>` のように `.` (ドット) でつなげて書くことでアクセスできる。  
\* パッケージ、モジュールはそれぞれフォルダ、ファイルにほぼ対応する

使い方や概要は `obj.capitalize??` や `help(obj.capitalize)` を実行する、または「\<パッケージ名\> \<関数名\>」などで検索することで調べられる。

````python
obj.capitalize??
````

````python
help(obj.capitalize)
````
